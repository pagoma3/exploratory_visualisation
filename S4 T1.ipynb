## Exercise 1

Make a statistical summary of the DelayedFlights.csv dataset.

Create at least a visualisation for:

- A categorical variable (UniqueCarrier)
- A numerical variable (ArrDelay)
- A numerical and a categorical variable (ArrDelay and UniqueCarrier)
- Two numerical variables (ArrDelay and DepDelay)
- Three variables (ArrDelay, DepDelay and UniqueCarrier)
- More than three variables (ArrDelay, DepDelay, AirTime and UniqueCarrier)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dades = pd.read_csv("Nou Fitxer")

del dades["Unnamed: 0"]

dades.isnull().sum()
```

```python
# UniqueCarrier visualisation:

plt.figure(figsize=[15, 10])

x = dades["UniqueCarrier"].value_counts()
noms = dades["UniqueCarrier"].value_counts().index.values

plt.pie(x, labels = noms)
plt.savefig("Pie_Airlines.pdf")
plt.show()
```

```python
# ArrDelay: numerical variable

plt.figure(figsize=[15, 10])

x = dades["ArrDelay"].value_counts()

plt.hist(x, range= (x.min(), 2400))     # Establish a limit of 24 hours of delay
plt.savefig("Bars_Delay.pdf")
plt.show()
```

```python
# A numerical and categorical variable (ArrDelay and UniqueCarrier)

grouped = dades[["ArrDelay", "UniqueCarrier"]] 

aseaet = grouped.groupby("UniqueCarrier").mean()

aseaet.sort_values(by = "ArrDelay", ascending = False)
 
################################################################    

plt.figure(figsize=(15, 10))

x = aseaet.index
y = aseaet["ArrDelay"]

plt.bar(x,y)
plt.savefig("Bars_Lines+Delay.pdf")
plt.show()

# This bar graph shows average delay per airline
# B6 and YV are some of the most delayed airlines; AQ and F9 are some of the most punctual
```

```python
# Two numerical variables: ArrDelay and DepDelay

plt.figure(figsize=(15, 10))

x = dades["DepDelay"]
y = dades["ArrDelay"]

plt.scatter(x,y) 
plt.savefig("Scatter_Delays.pdf")
plt.show()
```

```python
# Three variables: ArrDelay, DepDelay and UniqueCarrier

grouped = dades[["DepDelay", "ArrDelay", "UniqueCarrier"]]

ordenats = grouped.groupby("UniqueCarrier").mean()

ordenats.sort_values(by = "ArrDelay", ascending = False)
 
################################################################    

plt.figure(figsize=(15, 10))

width = 0.35

x = ordenats.index
y1 = ordenats["DepDelay"]
y2 = ordenats["ArrDelay"]

plt.bar(x, y1)
plt.bar(x, y2, bottom = y1) # Bottom to pile them
plt.savefig("Bars_Delays+Lines.pdf")
plt.show()

# This graph shows Departure and Arrival Delay in terms of airlines; no significant difference can be seen between
# Arrival and Departure delays in any case
```

```python
# More than three variables (ArrDelay, DepDelay, AirTime and UniqueCarrier)


def crearscatter(param):        # Scatterplot-creating function
    plt.figure(figsize=(15, 10))

    x = param["AirTime"]
    y1 = param["DepDelay"] 
    y2 = param["ArrDelay"] 
    
    plt.scatter(x, y1, marker = 'o')
    plt.scatter(x, y2, marker = '*')

    plt.xlabel("Flight length")
    plt.ylabel("Delay")
    
    plt.legend(["Departure Delay", "Arrival Delay"])

################################################################

# Intersting variables are kept:
    
grouped = dades[["DepDelay", "ArrDelay", "AirTime", "UniqueCarrier"]]

grouped.sort_values(by = "AirTime", ascending = True)

################################################################ 

# American Airlines flights delay (a lot of flights, moderate average delay)

american = grouped[grouped["UniqueCarrier"] == "AA"]

#american.head(10) 

crearscatter(american)

plt.title("American Airlines Delays")

plt.savefig("American_Delays.pdf")
plt.show()
 
```

```python
# Alaskan Airlines Analysis (Few flights, few delays)

alaskan = grouped[grouped["UniqueCarrier"] == "AQ"]

#alaskan.head(10)

crearscatter(alaskan)

plt.title("Alaskan Airlines Delays")

plt.savefig("Alaskan_Delays.pdf")
plt.show()
```

```python
# JetBlue Analysis (Few flights, few delays)

jetblue = grouped[grouped["UniqueCarrier"] == "B6"]

#jetblue.head(10)

crearscatter(jetblue)

plt.title("JetBlue Delays")

plt.savefig("JetBlue_Delays.pdf")
plt.show()
```
